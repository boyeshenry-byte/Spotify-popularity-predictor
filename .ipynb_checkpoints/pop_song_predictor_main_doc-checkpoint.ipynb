### Explanation of Variables:

The data set contains 32,833 songs with 14 features:
* `track_album_release_date` - The day month and year a track was released

* `playlist_genre` - The genre of songs contained within a playlist

* `danceability` - A quantitative feature between 0.0 and 1.0 to describe how suitable a song is for dancing based on its musical characteristics.

* `energy` - A perceptual measure from 0.0 to 1.0 of the intensity and activity of a song based on a song's characteristics. High energy tracks feel fast and loud (i.e. death metal).

* `key` - The musical scale a song is composed in.

* `loudness` - A measure of the overall loudness of a track in decibels (dB) averaged across an entire track.

* `mode` - Indicates the modality (major or minor) of a song. Either 1 (major) or 0 (minor).

* `speechiness` - The presence of spoken words in a track. The closer to 1, the more a track is like spoken words or poetry. Values above 0.66 are likely to be all spoken word while tracks between 0.33 and 0.66 are a mixed of music and speech. Values below 0.33 most likely represent music and other non-speechlike tracks.

* `acousticness` - A measure of confidence between 0.0 and 1.0 of how acoustic the track is with 1.0 being a high confidence. 

* `instrumentalness` - Whether or not a track contains vocals. The closer to 1, the greater the likelihood the track conatins no vocals.

* `liveness` - Whether an audience is present in the recording. Higher values (above 0.8) increase the probability a track was recorded live. 

* `valence` - A measure of the musical positiveness of a track between 0.0 and 1.0. Higher valence tracks sound more positive while a lower valence may sound more sad, depressed or angry.

* `tempo` - The overall estimated tempo in beats per minute (BPM).

* `duration_ms` - Duration of the track in milliseconds.


*SOURCE* developer.spotify.com/documentation/web-api/reference/get-audio-features

## KMeans clustering

```python
input_feat = log_df.select_dtypes('number').drop(columns=['pop_log']).copy()

X = StandardScaler().fit_transform(input_feat)
## Because we are already using the log adjusted variables, only standardizing now

clusters_2 = KMeans(n_clusters=2, random_state=121, n_init=25, max_iter=500).fit_predict(X)

log_copy = log_df.copy()

log_copy['k2'] = pd.Series(clusters_2, index=log_copy.index).astype('category')
```

```python
log_copy.groupby('k2').mean(numeric_only=True)
```

```python
%%time

tots_within = []

K= range(1, 21)

for _ in K:
    km = KMeans(n_clusters=_, random_state=121, n_init=25, max_iter=25)
    km = km.fit(X)

    tots_within.append(km.inertia_)
```

```python
fig, ax = plt.subplots()

ax.plot(K, tots_within, 'bo-')
ax.set_xlabel('number clusters')
ax.set_ylabel('total within sum of squares')
ax.set_xticks(range(len(K)+1))

plt.title('Optimal clusters for total within sum of squares')
plt.show()
```

The optimal number of clusters appears to be k=6. The total within sum of squares decreases sharply until this point, then tapers off.

```python
formula_list = ['y~1', # No INPUTS! Fit an INTERCEPT-ONLY or CONSTANT AVERAGE model.
               'y~x01 + x02 + x03 + x05 + x06', # Categorical inputs with additive features.
               'y~x07 + x08 + x09 + x10 + x11 + x12 + x13 + x14 + x15 + x16', # Continuous inputs with linear additive features.
               'y~x01 + x02 + x03 + x05 + x06 + x07 + x08 + x09 + x10 + x11 + x12 + x13 + x14 + x15 + x16', # All inputs (continuous and categorical) with linear additive features.
               'y~(x07 + x08 + x09 + x10 + x11 + x12 + x13 + x14 + x15 + x16)**2', # Continuous inputs with linear main effect and pair-wise interactions.
               'y~(x01 + x02 + x03 + x05 + x06 + x07 + x08 + x09 + x10 + x11 + x12 + x13 + x14 + x15 + x16)**2', # Interact the categorical inputs with the continuous inputs. This model must include the linear main effects as well
               'y~x01 + x02 + x03 + x05 + x06 + x07 + x08 + x09 + x10 + x11 + x12 + x13 + x14 + x15 + x16 + \
                np.power(x09, 2) + np.power(x12, 2) + np.sqrt(x13) + np.sqrt(x14) + np.power(x15, 2) + np.sqrt(x16)', # All inputs, with linear additive features and transformed variables
                'y~((x01 + x02 + x03 + x05 + x06) + (x07 + x08 + x09 + x10 + x11 + x12 + x13 + x14 + x15 + x16)**2) + \
               (np.power(x07, 2) + np.power(x08, 2) + np.power(x09, 2) + np.power(x10, 2) + np.power(x11, 2) + \
               np.power(x12, 2) + np.sqrt(x13) + np.sqrt(x14) + np.power(x15, 2) + np.sqrt(x16))' # All cat plus pairwise cont and derived features
               ]
```

### Equation selection

For this project I selected increasingly complex equations starting with a baseline, intercept only model.

Next was a model featuring only the additive effects of categorical inputs.

I followed this up with looking at only main effects of the continuous inputs.

Then we investigated all main effects of all inputs, followed by the main and interactive effects of all continuous inputs.

After those models, I tried a model featuring the interactions and main effects of all inputs, to try and derive any important features with dependent interactions with the intent of performing elastic net as an extenstion to this project if predictive power is low. 

The EDA showed that some of the input variables had a logarithmic relationship to the output. Because of this I decided to transform them using their square root. These variables are also the ones that we needed to log-shift for the exploration with the exception of instrumentalness, which appeared to have a somewhat parabolic relationship with the output. Some of the other variables appeared to have a parabolic relationship with the output. For these variables I derived new coefficients using their second power. 

Throughout the exploration, inputs appeared to have additive effects with each other, however, none seemed to create a strong influence over the output. I reasoned that these features likely had some form of interaction with each other. Thus, I decieded to add the derived coefficients to the original variables. 

I also decided to check if there were any feature interactions. To do this, I took the additive effects of the categorical variables in addition to the main effects and interactions of the continuous variables. I also added the derived variables to this equation as well.  

```python
kf = KFold(n_splits=5, shuffle=True, random_state=121)
```

```python
def fit_and_assess_ols(mod_name, a_formula, the_data):
    a_mod = smf.ols(formula=a_formula, data=the_data).fit()
    
    res_dict = {'model_name': mod_name,
                'model_formula': a_formula,
                'num_coefs': len(a_mod.params),
                'R-squared': a_mod.rsquared,
                'RMSE': np.sqrt((a_mod.resid **2 ).mean())}
    
    return pd.DataFrame(res_dict, index=[0])
```

```python
def mod_sum_stats(mod, formula):

    df = pd.DataFrame({
        'model_name': mod,
        'model_formula': formula,
        'coef_name': mod.params.index,
        'coef_value': mod.params.values,
        'std_err': mod.bse.values,
        'p_value': mod.pvalues,
        'significant': mod.pvalues < 0.05,
        '+/-': ['+' if val > 0 else 0 if val == 0 else '-' for val in mod.params.values]
    })

    df = df.reindex(df.coef_value.abs().sort_values(ascending=False).index)
    
    return df
```

```python
%%time 

ols_res_list = []

for m in range(len(formula_list)):
    print('-- Formula ID %d --' % m)

    try:
        ols_res_list.append(fit_and_assess_ols( m, formula_list[m], lm_df))
    except:
        print('!!!Formula ID %d could not fit!!!' % m)
```

```python
ols_res_df = pd.concat(ols_res_list)
```

While exploring this data, I found that the additive linear features had about as much predictive power as the more complex models with features derived based on the input/output relationship. With this in mind, I revised the formulas to best predict the data. Unfortunately to get any R-squared values greater than 0.08, this data set requires a lot of overfit, suggesting that as we saw in the exploration, it is a very complex dataset with many non-linear features. 

The model 7 is the linear additive features plus the derived features. This shows a small increase in performance, though not significant (0.01). It does show that there may be some influence of these feautres. This model was chosen because of the lack of any linear relationships during exploration. 

Model 8 is a model that features the linear additive categories with the pairwise continuous features. These are then added to the derived features. This model provides a balance of performance based on the additive and derived features, without creating too many coefficients and overfitting. 

```python
%%time

mod_intercept_only = smf.ols(formula=formula_list[0], data=lm_df).fit()  # y~1
mod_cat = smf.ols(formula=formula_list[1], data=lm_df).fit() # y~x01 + x02 + x03 + x05 + x06
mod_cont = smf.ols(formula=formula_list[2], data=lm_df).fit()      # y~x07 + ... + x16
mod_all_linear = smf.ols(formula=formula_list[3], data=lm_df).fit()      # y~all inputs linear
mod_cont_int = smf.ols(formula=formula_list[4], data=lm_df).fit()  # y~(x07 + ... + x16)**2
mod_cat_cont_int = smf.ols(formula=formula_list[5], data=lm_df).fit()  # y~(x01 + ... + x16)**2
mod_cont_feat = smf.ols(formula=formula_list[6], data=lm_df).fit()     # additive derived features
mod_cat_cont_feat_int = smf.ols(formula=formula_list[7], data=lm_df).fit()  # cont + pairwise : derived feat
```

```python
def sig_coefs(mod, formula):
    sig_sum = mod_sum_stats(mod, formula)
    
    sig_sum = sig_sum[sig_sum['significant']==True].reset_index()
    
    return sig_sum

sig_coefs(mod_all_linear, formula_list[3])
```

```python
kf = KFold(n_splits=5, random_state=121, shuffle=True)
```

```python
sk_lm = LinearRegression()
```

```python
def lm_cross_val_score(mod_name, a_formula, init_mod, the_data, cv):
    # create the feature and output arrays using the formula
    y, X = dmatrices( a_formula, data=the_data )

    # pipeline for preprocessing
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', init_mod)
    ])
    
    # train and test within each fold - return the test set scores
    # rsquared
    test_r2 = cross_val_score(pipe, X, y.ravel(), cv=kf, scoring='r2')
    # rmse
    test_rmse = -cross_val_score(pipe, X, y.ravel(), cv=kf, scoring='neg_root_mean_squared_error')
    
    # book keeping
    res_df=pd.DataFrame({'R-squared': test_r2,
                         'RMSE': test_rmse})
    
    res_df['fold_id'] = res_df.index + 1
    res_df['model_name'] = mod_name + 1
    res_df['model_formula'] = a_formula
    res_df['num_coefs'] = X.shape[1]
    
    return res_df
```

```python
%%time

cv_form_list = [3, 5, 7]

cv_score_list = []

for m in cv_form_list:
    cv_score_list.append(lm_cross_val_score(m, a_formula=formula_list[m], init_mod=sk_lm, the_data=lm_df, cv=kf))
```
