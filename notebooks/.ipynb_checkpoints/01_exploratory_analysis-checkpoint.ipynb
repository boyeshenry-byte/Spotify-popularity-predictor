## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore', category=UserWarning)

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## Data Overview

### Load Data

```python
songs_url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv'
songs_df = pd.read_csv(songs_url)

print(f"Dataset Shape: {songs_df.shape}")
songs_df.info()
```

### Variable Descriptions

**Target Variable:**
- `track_popularity` (0-100): Spotify's proprietary popularity metric

**Audio Features:**
- `danceability` (0-1): Suitability for dancing based on tempo, rhythm stability, beat strength
- `energy` (0-1): Perceptual intensity and activity (e.g., death metal = high energy)
- `loudness` (dB): Overall loudness averaged across entire track
- `speechiness` (0-1): Presence of spoken words (>0.66 = spoken word, <0.33 = music)
- `acousticness` (0-1): Confidence measure of acoustic instrumentation
- `instrumentalness` (0-1): Predicts lack of vocals (>0.5 = likely instrumental)
- `liveness` (0-1): Presence of audience (>0.8 = likely live recording)
- `valence` (0-1): Musical positiveness (high = happy, low = sad/angry)
- `tempo` (BPM): Beats per minute
- `duration_ms`: Track length in milliseconds

**Categorical Features:**
- `playlist_genre`: Top-level genre (edm, latin, pop, r&b, rap, rock)
- `key` (0-11): Musical key (0 = C, 1 = C♯/D♭, etc.)
- `mode` (0/1): Modality (1 = major, 0 = minor)

*Source: [Spotify Web API Documentation](https://developer.spotify.com/documentation/web-api/reference/get-audio-features)*

## Data Cleaning

### Check for Missing Values and Duplicates

```python
print("Missing Values:")
print(songs_df.isna().sum()[songs_df.isna().sum() > 0])

print(f"\nUnique tracks: {songs_df['track_id'].nunique()}")
print(f"Total rows: {len(songs_df)}")
print(f"Duplicates: {len(songs_df) - songs_df['track_id'].nunique()}")
```

### Remove Duplicates and Create Clean Dataset

```python
# Remove all duplicate track_ids (keeps none to avoid bias)
songs_no_dupe = songs_df.drop_duplicates(subset='track_id', keep=False).reset_index(drop=True)

# Convert categorical variables
songs_no_dupe['key'] = songs_no_dupe['key'].astype('category')
songs_no_dupe['mode'] = songs_no_dupe['mode'].astype('category')
songs_no_dupe['playlist_genre'] = songs_no_dupe['playlist_genre'].astype('category')

# Parse dates and extract temporal features
songs_no_dupe['track_album_release_date_dt'] = pd.to_datetime(
    songs_no_dupe['track_album_release_date'], 
    format='mixed'
)
songs_no_dupe['release_year'] = songs_no_dupe['track_album_release_date_dt'].dt.year.astype('category')
songs_no_dupe['release_month'] = songs_no_dupe['track_album_release_date_dt'].dt.month.astype('category')
songs_no_dupe['decades'] = ((songs_no_dupe['release_year'].astype(int) // 10) * 10).astype('category')

print(f"Clean dataset shape: {songs_no_dupe.shape}")
```

### Transform Target Variable

**Why transform popularity?**
Popularity is heavily right-skewed with many zeros. We apply a logit transformation to normalize the distribution for linear modeling.

```python
# Visualize original distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].hist(songs_no_dupe['track_popularity'], bins=50, edgecolor='black')
axes[0].set_title('Original Popularity Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Popularity Score (0-100)')
axes[0].set_ylabel('Frequency')

# Log transformation (shift 0s and 100s to avoid infinity)
songs_no_dupe['track_pop_shift'] = songs_no_dupe['track_popularity'].copy()
songs_no_dupe.loc[songs_no_dupe['track_popularity'] == 100, 'track_pop_shift'] = 99.9
songs_no_dupe.loc[songs_no_dupe['track_popularity'] == 0, 'track_pop_shift'] = 0.1

# Logit transform: log(p/(1-p))
songs_no_dupe['pop_frac'] = songs_no_dupe['track_pop_shift'] / 100
songs_no_dupe['pop_log'] = np.log(songs_no_dupe['pop_frac'] / (1 - songs_no_dupe['pop_frac']))

axes[1].hist(songs_no_dupe['pop_log'], bins=50, edgecolor='black', color='coral')
axes[1].set_title('Log-Transformed Popularity', fontsize=14, fontweight='bold')
axes[1].set_xlabel('log(p/(1-p))')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Create final Clean Dataset

```python
# Select relevant columns only
clean_df = songs_no_dupe[[
    'playlist_genre', 'key', 'mode', 'release_year', 'release_month', 'decades',
    'danceability', 'energy', 'loudness', 'valence', 'tempo', 'duration_ms',
    'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'pop_log'
]].copy()

# Apply log1p transform to highly skewed features
skewed_features = ['speechiness', 'acousticness', 'instrumentalness', 'liveness']
for feature in skewed_features:
    clean_df[f'{feature}_log'] = np.log1p(clean_df[feature])

# Create version with log-transformed features
log_df = clean_df.drop(columns=skewed_features).copy()

print(f"Final clean dataset: {clean_df.shape}")
print(f"Log-transformed dataset: {log_df.shape}")
```

## Distribution Analysis

### Categorical Features

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Genre
genre_counts = clean_df['playlist_genre'].value_counts()
axes[0].bar(genre_counts.index, genre_counts.values, color='steelblue', edgecolor='black')
axes[0].set_title('Song Distribution by Genre', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Count')
for i, (genre, count) in enumerate(genre_counts.items()):
    axes[0].text(i, count + 100, str(count), ha='center', va='bottom')

# Decades
decade_counts = clean_df['decades'].value_counts().sort_index()
axes[1].bar(decade_counts.index.astype(str), decade_counts.values, color='coral', edgecolor='black')
axes[1].set_title('Song Distribution by Decade', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=45)

# Key
key_counts = clean_df['key'].value_counts().sort_index()
axes[2].bar(key_counts.index.astype(str), key_counts.values, color='lightgreen', edgecolor='black')
axes[2].set_title('Song Distribution by Key', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Count')

# Mode
mode_counts = clean_df['mode'].value_counts()
mode_labels = ['Minor (0)', 'Major (1)']
axes[3].bar(mode_labels, mode_counts.values, color='plum', edgecolor='black')
axes[3].set_title('Song Distribution by Mode', fontsize=12, fontweight='bold')
axes[3].set_ylabel('Count')
for i, count in enumerate(mode_counts.values):
    axes[3].text(i, count + 500, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

**Key Observations:**
- Rock dominates the dataset (6,950 songs)
- Most songs released in 2010s (17,564 songs) - dataset has recency bias
- Songs fairly evenly distributed across keys
- Major mode (~65%) more common than minor

### Continuous Features

```python
# Select numeric features for visualization
numeric_features = [
    'danceability', 'energy', 'loudness', 'valence', 'tempo', 
    'duration_ms', 'speechiness', 'acousticness', 'instrumentalness', 'liveness'
]

fig, axes = plt.subplots(3, 4, figsize=(16, 10))
axes = axes.flatten()

for i, feature in enumerate(numeric_features):
    axes[i].hist(clean_df[feature], bins=40, edgecolor='black', alpha=0.7)
    axes[i].set_title(feature.replace('_', ' ').title(), fontsize=10, fontweight='bold')
    axes[i].set_ylabel('Frequency')

# Hide extra subplots
for i in range(len(numeric_features), len(axes)):
    axes[i].axis('off')

plt.suptitle('Distribution of Audio Features', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()
```

**Key Observations:**
- Strong right skew in: speechiness, acousticness, instrumentalness, liveness
- Relatively uniform distributions: danceability, energy, valence
- Loudness approximately normal (centered around -6 dB)

## Correlation Analysis

```python
# Calculate correlation matrix for log-transformed features
corr_matrix = log_df.select_dtypes(include='number').corr()

fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    fmt='.2f',
    cmap='coolwarm', 
    center=0,
    vmin=-1, 
    vmax=1,
    square=True,
    cbar_kws={'label': 'Correlation'},
    ax=ax
)
plt.title('Correlation Heatmap: Audio Features vs Popularity', 
          fontsize=14, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Print strongest correlations with popularity
pop_corr = corr_matrix['pop_log'].sort_values(ascending=False)
print("\nCorrelations with Popularity (pop_log):")
print(pop_corr[pop_corr.index != 'pop_log'])
```

**Critical Finding:**
> **No audio feature strongly correlates with popularity.** The strongest correlation is acousticness (+0.09) and energy (-0.10). This suggests audio features alone cannot predict song popularity.

**Notable feature correlations:**
- Energy ↔ Loudness: **+0.76** (loud songs are energetic)
- Energy ↔ Acousticness: **-0.69** (acoustic songs are less energetic)
- Valence ↔ Danceability: **+0.48** (happy songs are more danceable)

## Feature Relationships

### Audio Features vs Popularity

```python
# Select 3 most correlated features + 3 interpretable features
features_to_plot = ['acousticness_log', 'energy', 'danceability', 
                    'loudness', 'valence', 'duration_ms']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, feature in enumerate(features_to_plot):
    axes[i].scatter(
        log_df[feature] if feature.endswith('_log') else clean_df[feature],
        clean_df['pop_log'],
        alpha=0.3,
        s=10
    )
    
    # Add trendline
    x = log_df[feature].values if feature.endswith('_log') else clean_df[feature].values
    y = clean_df['pop_log'].values
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    axes[i].plot(x, p(x), "r-", linewidth=2, label=f'y={z[0]:.3f}x+{z[1]:.2f}')
    
    axes[i].set_xlabel(feature.replace('_', ' ').title(), fontsize=10)
    axes[i].set_ylabel('Log Popularity', fontsize=10)
    axes[i].set_title(f'{feature.replace("_", " ").title()} vs Popularity', 
                      fontsize=11, fontweight='bold')
    axes[i].legend(loc='best', fontsize=8)
    axes[i].grid(alpha=0.3)

plt.suptitle('Relationship Between Audio Features and Popularity', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Observations:**
- All relationships show **high scatter** and **weak trends**
- Energy: slight negative trend (r = -0.10)
- Danceability: slight positive trend (r = +0.05)
- No feature shows strong predictive power

### Categorical Features vs Popularity

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Genre
sns.boxplot(
    data=clean_df, 
    x='playlist_genre', 
    y='pop_log',
    palette='Set2',
    ax=axes[0]
)
axes[0].set_title('Popularity by Genre', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Genre', fontsize=10)
axes[0].set_ylabel('Log Popularity', fontsize=10)

# Decade
decade_order = sorted(clean_df['decades'].unique())
sns.boxplot(
    data=clean_df, 
    x='decades', 
    y='pop_log',
    order=decade_order,
    palette='viridis',
    ax=axes[1]
)
axes[1].set_title('Popularity by Decade', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Decade', fontsize=10)
axes[1].set_ylabel('Log Popularity', fontsize=10)
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Calculate mean popularity by genre
genre_stats = clean_df.groupby('playlist_genre')['pop_log'].agg(['mean', 'median', 'std'])
genre_stats = genre_stats.sort_values('mean', ascending=False)
print("\nPopularity Statistics by Genre:")
print(genre_stats)
```

**Findings:**
- Genre shows **modest differences** in popularity:
  - Rock, Pop, Rap: higher average popularity
  - R&B, EDM: lower average popularity
  - High overlap in IQRs suggests genre alone is not decisive
- Decade shows some variation, but outliers dominate each category

## Key Findings

### Summary of EDA

1. **Dataset Characteristics:**
   - 25,190 unique songs across 6 genres
   - Heavy recency bias (70% from 2010s)
   - Popularity heavily right-skewed → requires log transformation

2. **Weak Feature Relationships:**
   - **NO audio feature correlates strongly with popularity** (all |r| < 0.10)
   - Strongest predictors: acousticness (+0.09), energy (-0.10)
   - High scatter in all relationships suggests non-linear patterns

3. **Categorical Insights:**
   - Genre shows slight popularity differences (Rock/Pop/Rap > EDM/R&B)
   - Key and mode show minimal popularity variation
   - Decade effects present but confounded by data collection bias

4. **Implications for Modeling:**
   - Audio features alone likely insufficient for strong predictions
   - Expect low R² values (<0.15) in linear models
   - Missing critical factors: artist fame, marketing, playlist placement, algorithm boosting
